**REMOVE DUPLICATES**

```python
customers = customers.drop_duplicates()
accounts = accounts.drop_duplicates()
transactions = transactions.drop_duplicates()
loans = loans.drop_duplicates()
```

```python
customers.duplicated().sum()
```

**HANDLE MISSING VALUES**

```python
customers.isnull().sum()
```

Fill the missing occupation, gender, and city.

```python
columns_to_fill = ["occupation", "city", "gender"]
customers[columns_to_fill] = customers[columns_to_fill].fillna("Unknown")
```

```python
customers.head(15)
```

Customer ID should never be null. The following code drops records with missing critical IDs.

```python
customers = customers.dropna(subset=["customer_id"])
```

Convert the dates from object to date type

```python
customers["join_date"] = pd.to_datetime(customers["join_date"])
transactions["transaction_date"] = pd.to_datetime(transactions["transaction_date"])
```

**HANDLE INCONSISTENT FORMATTING**

```python
customers["gender"] =customers["gender"].str.title()
customers["occupation"] =customers["occupation"].str.title()
customers["city"] = customers["city"].str.title()
accounts["branch"] = accounts["branch"].str.title()
accounts["account_type"] = accounts["account_type"].str.title()
accounts["account_status"] = accounts["account_status"].str.title()
loans["loan_type"] = loans["loan_type"].str.title()
transactions["transaction_type"] = transactions["transaction_type"].str.title()
transactions["transaction_status"] = transactions["transaction_status"].str.title()
transactions["transaction_channel"] = transactions["transaction_channel"].str.upper()
customers["email"] = customers["email"].str.strip()
```

VALIDATE RELATIONSHIPS AND IDs

```python
invalid_transactions = transactions[
    ~transactions["customer_id"].isin(customers["customer_id"])
]

print(invalid_transactions)
```

**HANDLE OUTLIERS**
Remove negative transactions

```python
transactions = transactions[
    transactions["transaction_amount"] > 0
]
```

```python
transactions.describe()
```

**DETECT ANOMALIES**

```python
threshold = transactions["transaction_amount"].quantile(0.99)

anomalies = transactions[
    transactions["transaction_amount"] > threshold
]
```

**DERIVE COLUMNS**:
Customer Age


```python
today = pd.Timestamp.today()
customers["age"] = (
    today.year - pd.to_datetime(customers["date_of_birth"]).dt.year
)
```
